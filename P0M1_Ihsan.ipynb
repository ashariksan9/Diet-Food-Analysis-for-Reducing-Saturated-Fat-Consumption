# Data Loading & Data Cleaning

Setelah kita mengetahui permasalahan yang ingin kita analisis, kita akan melakukan data loading yang mana selanjutnya dilakukan pengecekan apakah data tersebut masih raw atau sudah bersih agar dapat dilakukan data cleaning.

```python
import pandas as pd

"""

Dataset yangk kita miliki diambil dari kaggle dan didownload dalam bentuk file csv. Datasets dapat diakses melalui url berikut: 
`https://www.kaggle.com/datasets/utsavdey1410/food-nutrition-dataset/data`

"""

df1 = pd.read_csv('FOOD-DATA-GROUP1.csv')
df2 = pd.read_csv('FOOD-DATA-GROUP2.csv')
df3 = pd.read_csv('FOOD-DATA-GROUP3.csv')

df = pd.concat([df1, df2, df3], axis=0)

df.reset_index(drop=True, inplace=True)

df
```

```python
df.info()
```

Dari informasi yang ditampilkan terlihat sepertinya data dalam keadaan cukup bersih dan juga memiliki data type yang benar.

```python
df.isna().sum()
```

Setelah pengecekan missing value juga tetap sama, tidak ada data yang berisi missing value.

```python
df.columns
```

Pengecekan selanjutnya diatas melihat apakah terdapat whitespace pada kolom, dan ternyata juga sudah cukup bersih.

```python
df[df.duplicated()]
```

Saat dilakukan pengecekan duplicate data juga tidak terlihat adanya duplikasi diatara row dataframe tersebut.

Dengan begitu selanjutkan kita akan melanjutkan dengan mengambil kolom kolom yang ingin kita gunakan saja.

```python
# Daftar kolom yang ingin dipilih
kolom_perlu = [
    'food',
    'Saturated Fats',
    'Nutrition Density',
    'Protein',
    'Dietary Fiber',
    'Caloric Value',
    'Monounsaturated Fats',
    'Polyunsaturated Fats',
    'Carbohydrates'
]

# Memilih kolom yang diperlukan dan membuat DataFrame baru
food_data = df[kolom_perlu]

food_data
```

Melihat dari penjabaran masalah, terdapat 9 kolom yang kita perlukan untuk mengalasis permasalahannya. Hasil pemisahan kolom-kolom tersebut dari dataframe awal dapat dilihat diatas.

# Analisis dan Perhitungan

Setelah kita mendapatkan data yang kita inginkan, selanjutnya masuk kepada tahapan analisis dan perhitungan. Didalam analasis kita akan mencoba menjawab ber urutan dari pertanyaan yang kita buat di penjabaran masalah.

Pertama kita akan mencari makanan apa saja yang memiliki kontribusi lemak jenuh yang besar, kita dapat melakukan analisis dengan membuat bar chart.

```python
import matplotlib.pyplot as plt

# menyatukan kolom makanan dan lemak jenuh untuk dihitung totalnya
top_saturated_fat = food_data.groupby('food')['Saturated Fats'].sum().sort_values(ascending=False)

# Visualisasi bar chart
plt.figure(figsize=(10, 6))
top_saturated_fat.head(10).plot(kind='bar', color='orange')
plt.title('Top 10 Makanan Dengan Lemak Jenuh Tinggi')
plt.xlabel('Makanan')
plt.ylabel('Total Lemak Jenuh (g)')
plt.show()
```

Dari chart kita dapat melihat 10 makanan teratas dengan lemak jenuh yang tinggi. Coconut (whole coconut) mendominasi chart dengan total lemak jenuh yang jauh lebih tinggi dibandingkan makanan lainnya. Ini menunjukkan bahwa kelapa utuh memiliki kontribusi terbesar terhadap konsumsi lemak jenuh dalam diet. Coconut whole dan Weetabix adalah dua makanan yang paling berkontribusi besar terhadap konsumsi lemak jenuh.

Selanjutnya kita akan melihat korelasi antara lemak jenuh dan juga kepadatan nutrisi didialamnya.

```python
# scatter plot melihat persebaran makanan yang memiliki lemak jenuh dan kepadatan nutrisi
plt.figure(figsize=(10, 6))
plt.scatter(food_data['Nutrition Density'], food_data['Saturated Fats'], color='red', alpha=0.5)
plt.title('Lemak Jenuh vs. Kepadatan Nutrisi')
plt.xlabel('Kepadatan Nutrisi')
plt.ylabel('Lemak Jenuh')
plt.grid(True)
plt.show()
```

Ada beberapa outlier dengan lemak jenuh yang sangat tinggi (misalnya lebih dari 50 gram) tetapi kepadatan nutrisinya relatif rendah. Titik-titik yang berada di bagian bawah kiri grafik menunjukkan makanan dengan kepadatan nutrisi yang rendah tetapi mengandung lemak jenuh yang tinggi. Makanan ini dapat dikategorikan sebagai makanan yang berpotensi tidak sehat. Variasi yang banyak ini cukup memberikan pengetahuan bahwa tidak ada korelasi yang jelas antara kepadatan nutrisi dan kandungan lemak jenuh. Artinya, makanan yang memiliki kepadatan nutrisi tinggi tidak selalu rendah dalam lemak jenuh dan sebaliknya.

Selanjutnya kita akan melihat apakah ada pola terhadap makanan yang memiliki lemak jenuh yang tinggi.

```python
plt.figure(figsize=(10, 6))
plt.hist(food_data['Saturated Fats'], bins=30, color='blue', edgecolor='black')
plt.title('Distribuasi Lemak Jenuh pada Makanan')
plt.xlabel('Lemak Jenuh (g)')
plt.ylabel('Frequency')
plt.show()
```

Nah bisa kita lihat dari histogram diatas dari dalam dataset yang kita miliki cenderung makanan yang memiliki lemak jenuh yang rendah. bisa dilihat frekuensi nya sangat tinggi di bagian awal yang hampir menyentuk 1400. Dengan begitu, ada banyak opsi makanan yang bisa kita pakai dalam menunjang program diet kita.

Selanjutnya untuk analisis lebih dalam, kita ingin membandingkan rasio lemak jenuh terhadap protein dan juga serat pada makanan.

```python
import seaborn as sns
import numpy as np

food_data['Rasio Lemak Jenuh ke Protein'] = food_data['Saturated Fats'] / food_data['Protein']
food_data['Rasio Lemak Jenuh ke Serat'] = np.where(
    food_data['Dietary Fiber'] == 0,
    np.nan, 
    food_data['Saturated Fats'] / food_data['Dietary Fiber']
) # pengubahan data 0 ke nan dikarenakan ada beberapa makanan yang tidak memiliki serat sehingga dapat menimbulkan nilai inf jika dihitung


plt.figure(figsize=(10, 6))
sns.boxplot(data=food_data[['Rasio Lemak Jenuh ke Protein', 'Rasio Lemak Jenuh ke Serat']])
plt.title('Distribusi Rasio Lemak Jenuh ke Protein dan Serat')
plt.show()
```

Grafik menunjukkan distribusi rasio antara lemak jenuh dengan protein serta lemak jenuh dengan serat. Dari visualisasi ini, kita dapat melihat bahwa sebagian besar nilai rasio ini cenderung mendekati 0, yang menunjukkan bahwa pada umumnya makanan dalam dataset memiliki lebih sedikit lemak jenuh dibandingkan dengan protein atau serat.

Dilihat dari bentuk dan posisi boxplot, rasio lemak jenuh terhadap serat tampaknya sedikit lebih terdistribusi dibandingkan dengan rasio lemak jenuh terhadap protein. Ini bisa mengindikasikan bahwa makanan dalam dataset lebih bervariasi dalam hal kandungan seratnya dibandingkan kandungan proteinnya.

Dengan begitu, selanjutnya kita akan melihat kapan makanan tersebut memiliki lemak yang cukup tinggi.

```python
"""
Kita menggunakan referensi dari websita siloam hospitals yang dapat diakses melalui 
url:https://www.siloamhospitals.com/informasi-siloam/artikel/mengenal-lemak-jenuh.
Dari web tersebut didapatkan normalnya kita mengkonsumsi lemak jenuh berada di angka 13 gram per hari.
kita akan mengkategorikan makanan yang memiliki lemak jenuh sebesar 10 gram ke dalam kategori tinggi lemak.
"""

# memasukkan makanan dengan lemak jenuh lebih dari 10 gram ke variabel lemak_tinggi
lemak_tinggi = food_data[food_data['Saturated Fats'] > 10]
lemak_tinggi_count = lemak_tinggi.shape[0]

# melihat statistik lemak jenuh dalam dataset
mean_saturated_fat = lemak_tinggi['Saturated Fats'].mean()
print(f'Rata-rata Lemak Jenuh di Makanan Tinggi Lemak: {mean_saturated_fat:.2f} g')
print(f'Jumlah Makanan dengan Lemak Jenuh Tinggi: {lemak_tinggi_count}')
```

Rata-rata lemak jenuh sebesar 33.50 g di makanan tinggi lemak menunjukkan bahwa makanan-makanan ini secara umum memiliki kandungan lemak jenuh yang sangat tinggi. Jumlah makanan yang masuk ke dalam kategori tinggi lemak hanya sebanyak 158 makanan, hal ini menandakan banyaknya opsi yang kita miliki untuk dapat menunjang program diet yang akan digunakan nantinya.

Sebelum kita melanjutkan analisis, disini kita menambahkan kolom fat category untuk menandakan makanan mana yang tinggi lemak dan mana yang normal.

```python
# Menambahkan kolom Fat Category
food_data['Fat Category'] = food_data['Saturated Fats'].apply(lambda x: 'High Fat' if x > 10 else 'Normal Fat')

food_data.sample(20)
```

Dapat dilihat diatas kita telah membuat 3 kolom baru yaitu rasio lemak jenuh ke protein dan sert, dan juga kolom kategori fat. Dilihat juga bahwa sekarang beberapa row jadi memiliki value Nan yang mana akan kita ubah dulu ke 0. Cleaning dilakukan pada fase ini dikarenakan kolom tersebut baru timbul saat kita melakukan analisis.

```python
# Ganti NaN di kolom "Rasio Lemak Jenuh ke Protein" dengan 0
food_data['Rasio Lemak Jenuh ke Protein'] = food_data['Rasio Lemak Jenuh ke Protein'].fillna(0)

# Ganti NaN di kolom "Rasio Lemak Jenuh ke Serat" dengan 0
food_data['Rasio Lemak Jenuh ke Serat'] = food_data['Rasio Lemak Jenuh ke Serat'].fillna(0)

# Cek hasilnya
food_data[['Rasio Lemak Jenuh ke Protein', 'Rasio Lemak Jenuh ke Serat']].head(20)
```

Sekarang sudah terlihat yang tadinya missing value sudah berubah menjadi 0. Maka dari itu analisis bisa kita lanjutkan.

Terakhir kita akan membandingkan protein yang merupakan salah satu komponen penting dalam diet, dengan makanan kategori tinggi lemak dan lemak normal. Kita akan menggunakan analisis hipotesis yang mana akan melihat adakah perbedaan signifikan antara protein sebagai salah satu part penting dengan makanan yang memiliki kategori High Fat dan Normat Fat.

```python
from scipy.stats import ttest_ind
'''
Hipotesis Nol (H0): Tidak ada perbedaan rata-rata protein antara makanan dengan kategori High Fat dan Normal Fat.
Hipotesis Alternatif (H1): Ada perbedaan rata-rata protein antara makanan dengan kategori High Fat dan Normal Fat.
'''

# t-test untuk membandingkan Protein antara High Fat dan Normal Fat
high_fat = food_data[food_data['Fat Category'] == 'High Fat']['Protein']
normal_fat = food_data[food_data['Fat Category'] == 'Normal Fat']['Protein']
t_stat, p_value = ttest_ind(high_fat, normal_fat)

print(f't-statistics = {t_stat}, P-Value = {p_value}')
```

P-Value berada di angka 1.47e-72 adalah angka yang sangat kecil (mendekati nol) yang berarti bahwa perbedaan rata-rata kandungan protein antara dua kategori makanan ini cukup signifikan secara statistik. 

Jika rata-rata protein pada makanan lemak tinggi lebih rendah, ini bisa menjadi indikasi bahwa makanan-makanan ini tidak hanya tinggi lemak jenuh, tetapi juga mungkin kurang dalam kandungan protein, menjadikannya kurang ideal dari segi keseimbangan nutrisi.

Selanjutnya kita akan membuat sebuah filter yang mana akan kita gunakan untuk melihat makanan apa saja yang bisa kita gunakan dalam program diet.

```python
# Filter makanan yang masuk ke kategori "Normal Fat"
makanan_diet = food_data[food_data['Fat Category'] == 'Normal Fat']

# Kriteria untuk nutrisi tinggi
batas_protein = makanan_diet['Protein'].quantile(0.75)
batas_serat = makanan_diet['Dietary Fiber'].quantile(0.75)

# Filter makanan yang memenuhi syarat nutrisi tinggi
high_nutrient_foods = makanan_diet[
    (makanan_diet['Protein'] > batas_protein) |
    (makanan_diet['Dietary Fiber'] > batas_serat) 
]

total_makanan_diet = high_nutrient_foods['food'].count()

print(high_nutrient_foods[['food', 'Protein', 'Dietary Fiber']])
print(f"Jumlah makanan yang dapat digunakan dalam diet: {total_makanan_diet}")
```

Kita telah mendapatkan list makanan yang sehat untuk kita konsumsi untuk program diet kita. Perlu diketahui bahwa disini kita mengambil makanan sehat mengacu kepada website halodoc yang dapat di akses melalui url:https://www.halodoc.com/kesehatan/makanan-sehat?srsltid=AfmBOooRj4wj37auqmVzHpR5hUkNnMUA52hFC_6kdMttwWD1O7JYQoLo. Dimana kita mengambil makanan yang tinggi akan protein dan serat, tetapi juga rendah lemak jenuh.
